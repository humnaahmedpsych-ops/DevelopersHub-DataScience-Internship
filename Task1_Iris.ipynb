## Dataset Overview

The Iris dataset contains 150 observations and 5 columns. The dataset has no missing values, and the features include measurements of sepal length, sepal width, petal length, petal width, and the species of the iris flower.

## Scatter Plot

The scatter plot is used to visualize the relationship between petal length and petal width across different iris species.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species"
)

plt.title("Petal Length vs Petal Width")

plt.show()
```

## Observation

The scatter plot shows a positive relationship between petal length and petal width. Different iris species form distinct clusters, indicating that these features are useful for distinguishing between species.

## Histogram

A histogram is used to visualize the distribution of a numerical variable. It helps us understand how frequently different values occur in the dataset.

```python
plt.figure(figsize=(8,6))

sns.histplot(
    data=iris,
    x="sepal_length",
    bins=10,
    kde=True
)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

## Observation

The histogram shows the distribution of sepal length values in the Iris dataset. Most flowers have sepal lengths clustered around the middle of the distribution, while very small and very large values occur less frequently.

## Box Plot

A box plot is used to visualize the spread of numerical data and identify potential outliers. It displays the median, quartiles, and the overall distribution of the variable.

plt.figure(figsize=(8,6))

sns.boxplot(
    data=iris,
    y="sepal_length"
)

plt.title("Box Plot of Sepal Length")
plt.xlabel("Sepal Length")

plt.show()

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=iris,
    y="sepal_length"
)

plt.title("Box Plot of Sepal Length")
plt.xlabel("Sepal Length")

plt.show()
```

## Observation

The box plot summarizes the distribution of sepal length values. Most observations fall within a consistent range, and there are no significant outliers in the dataset. The median value is located near the center of the box, indicating a fairly balanced distribution.
